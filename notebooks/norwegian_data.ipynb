```python
data_daily = pd.read_csv("../data/fhi_data_daily.csv")
```

```python
data_daily.columns
```

```python
data_daily = data_daily[data_daily.year.astype(int) < 2021].reset_index()
data_daily = data_daily[['date', 'I_new', 'I_cumulative', 'D_new', 'D_cumulative']]
data_daily = data_daily['date'].apply(lambda x: 
                                        x.split("/")[1].zfill(2) + "/"
                                        + x.split("/")[0].zfill(2) + "/"
                                        + x.split("/")[2].zfill(2))
data_daily
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data_daily.date, data_daily.I_new, data_daily.I_cumulative)
plt.xticks(data_daily.date[::14], data_daily.date[::14], rotation='vertical')
plt.show()
```

```python
plt.plot(data_daily.date, data_daily.D_new, data_daily.D_cumulative)
plt.xticks(data_daily.date[::14], data_daily.date[::14], rotation='vertical')
plt.show()
```
